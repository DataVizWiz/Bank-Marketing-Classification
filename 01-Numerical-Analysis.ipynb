```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import viz

%matplotlib inline
sns.set_style('darkgrid')
sns.set_context('paper', font_scale=1.4)
```

```python
df = pd.read_csv('data/customers.csv', sep=';')
df.head()
```

```python
df_num = df[['age', 'balance', 'duration']]
df_num.head()
```

```python
class Histplot:
    
    def __init__(self, x, bins):
        self.x = x
        self.bins = bins
        self.df = df_num[[x]]
        
    def create_histplot(self):
        plt.figure(figsize=(18, 6))
        plt.title('Distribution of {}'.format(self.x))
        sns.histplot(self.df, kde=True, bins=self.bins);

num_list = df_num.columns

for col in num_list:
    histplot = Histplot(col, 50)
    histplot.create_histplot()
```

```python
sns.jointplot(x='age', y='balance', data=df, hue='education', height=12);
```

```python
sns.jointplot(x='age', y='balance', data=df, kind='kde', hue='marital', height=12);
```

```python
plt.figure(figsize=(18, 6))
plt.title('KDE Plot - Age')
sns.kdeplot(data=df, x='age', hue='housing', multiple='stack');
```

```python
sns.pairplot(df_num, palette='Blues', height=12);
```

```python
plt.figure(figsize=(18, 6))
plt.title('Rug Plot - Age')
sns.rugplot(df_num.age);
```

```python
plt.figure(figsize=(18, 6))
plt.title('Correlation between Numerical Variables')
plt.xticks(rotation=45)
sns.set_context('paper', font_scale=1.4)

sns.heatmap(df_num.corr(), annot=True, cmap='Blues');
```
