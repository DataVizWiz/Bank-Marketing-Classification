```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
sns.set_style('darkgrid')
sns.set_context('paper', font_scale=1.4)
```

```python
df = pd.read_csv('data/customers.csv', sep=';')
df.head()
```

```python
sns.displot(df['age'], kde=True, bins=30);
```

```python
sns.jointplot(x='age', y='balance', data=df, kind='reg');
```

```python
sns.jointplot(x='age', y='balance', data=df, kind='kde');
```

```python
sns.jointplot(x='age', y='balance', data=df, kind='hex');
```

```python
sns.kdeplot(df['age']);
```

```python
sns.pairplot(df, hue='y', palette='Blues');
```

```python
sns.rugplot(df['age'])
```
